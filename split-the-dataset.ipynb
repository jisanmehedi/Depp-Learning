```python
import os

os.listdir("/kaggle/working/Train_Test_Folder")
```

```python
train_dir = "/kaggle/working/Train_Test_Folder/train"
test_dir = "/kaggle/working/Train_Test_Folder/test"
val_dir = "/kaggle/working/Train_Test_Folder/val"
```

```python
training_images = []

for root, dirs, files in os.walk(train_dir):
    for file in files:
        training_images.append(os.path.join(root, file))
                               
validation_images = []

for root, dirs, files in os.walk(val_dir):
    for file in files:
        validation_images.append(os.path.join(root, file))
  
test_images = []

for root, dirs, files in os.walk(test_dir):
    for file in files:
        test_images.append(os.path.join(root, file))
```

```python
print("Training:")
print("Training Path:" + train_dir)
print("Training Classes:" + str(len(os.listdir(train_dir))))
print("Training Images:" + str(len(training_images)))

print("\n")

print("Validation:")
print("Validation Path:" + val_dir)
print("Validation Classes:" + str(len(os.listdir(val_dir))))
print("Validation Images:" + str(len(validation_images)))

print("\n")

print("Testing:")
print("Testing Path:" + test_dir)
print("Testing Classes:" + str(len(os.listdir(test_dir))))
print("Testing Images:" + str(len(test_images)))
```
